# **Exploring dataset and preprocessing**

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import imblearn
%matplotlib inline
```

```python
for col in dataset.columns:
    print(col)
```

```python
# drop connection and network dependant features
drop_columns = ['src_ip',	'dst_ip',	'src_port',	'dst_port',	'protocol',	'timestamp']
dataset = dataset.drop(columns=drop_columns)
```

```python
dataset['class'].value_counts()
```

```python
dataset.describe()
```

```python
# elimnate unchanged features
drop_columns = ['fwd_psh_flags',
                'bwd_psh_flags',
                'fwd_urg_flags',
                'bwd_urg_flags',
                'syn_flag_cnt',
                'rst_flag_cnt',
                'psh_flag_cnt',
                'ack_flag_cnt',
                'urg_flag_cnt',
                'ece_flag_cnt',
                'cwe_flag_count']
dataset = dataset.drop(columns=drop_columns)
```

```python
for col in dataset.columns:
    print(col)
```

```python
correlation_matrix = dataset.corr()
```

```python
correlation_matrix
```

```python
rows, cols = dataset.shape
flds = list(dataset.columns)
threshold = 0.95

# Indexing with numbers on a numpy matrix will probably be faster
corr = dataset.corr().values
drop_columns = []
for i in range(cols):
    for j in range(i+1, cols):
        if abs(corr[i,j]) > threshold:
            drop_columns.append(flds[j])
            print(flds[i], ' ', flds[j], ' ', corr[i,j])

drop_columns = list(set(drop_columns))
```

```python
drop_columns
```

```python
dataset = dataset.drop(columns=drop_columns)
```
